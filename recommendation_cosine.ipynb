```python
movie_ratings_pivot = movie_user_ratings.pivot_table(index='userId', columns='title', values='rating')
```

```python
movie_ratings_pivot = movie_ratings_pivot.dropna(thresh=10, axis=1).fillna(0)
```

```python
movie_ratings_pivot.head()
```

```python
def standardize(row):
    standard_row = (row-row.mean()) / (row.max()-row.min())
    return standard_row

movie_ratings_std = movie_ratings_pivot.apply(standardize)
movie_ratings_std.head()
```

```python
from sklearn.metrics.pairwise  import cosine_similarity

movie_similarity = cosine_similarity(movie_ratings_std.T)
movie_similarity
```

```python
movie_similarity_df = pd.DataFrame(movie_similarity, index=movie_ratings_std.columns, columns=movie_ratings_std.columns)
movie_similarity_df
```

```python
movie_similarity_df['10,000 BC (2008)']
```

```python
def get_recommendations(movie, rating, no_of_recomendations=10):
    mean_rating = movie_user_ratings['rating'].unique().mean()
    similarity_score = movie_similarity_df[movie] * (rating - mean_rating)
    similarity_score = similarity_score.sort_values(ascending=False).head(no_of_recomendations)
    return similarity_score
```

```python
get_recommendations('Avengers: Age of Ultron (2015)', 4)
```

```python
ratings_by_user1 = [('Avengers: Age of Ultron (2015)', 4), ('Captain America: The First Avenger (2011)', 5), ('Zombieland (2009)', 2), ('Wolf of Wall Street, The (2013)', 3)]
recommended_movies = pd.DataFrame()

for movie, rating in ratings_by_user1:
    recommended_movies = recommended_movies.append(get_recommendations(movie, rating), ignore_index=True)
    
recommended_movies.head()
```
